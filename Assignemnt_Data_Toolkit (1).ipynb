```python
# 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a DataFrame
data = {
    "Math": [80, 85, 90, 95],
    "Science": [78, 88, 92, 96],
    "English": [70, 75, 85, 90]
}

df = pd.DataFrame(data)

# Calculate correlation matrix
corr_matrix = df.corr()

# Plot heatmap
sns.heatmap(corr_matrix, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
#8. Create a line plot with multiple lines using Matplotlib
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y1 = np.array([2, 4, 6, 8, 10])
y2 = np.array([3, 6, 9, 12, 15])

plt.plot(x, y1, label='Line 1', color='blue', marker='o')
plt.plot(x, y2, label='Line 2', color='red', marker='s')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Line Plot')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 32, 18, 47, 29],
    'Salary': [50000, 60000, 30000, 80000, 55000]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

age_threshold = 30
filtered_df = df[df['Age'] > age_threshold]

print("\nFiltered DataFrame (Age > 30):")
print(filtered_df)
```

```python
#11. Perform matrix multiplication using NumPy
import numpy as np
import numpy as np

# Define two matrices
A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([[7, 8],
              [9, 10],
              [11, 12]])

# Method 1: Using np.dot()
C1 = np.dot(A, B)
print("Matrix multiplication using np.dot():\n", C1)

# Method 2: Using @ operator
C2 = A @ B
print("Matrix multiplication using @ operator:\n", C2)

# Method 3: Using np.matmul()
C3 = np.matmul(A, B)
print("Matrix multiplication using np.matmul():\n", C3)
```

```python
#13. Create a 3D scatter plot using Plotly.
import plotly.graph_objects as go
import numpy as np

np.random.seed(0)
x = np.random.randn(50)
y = np.random.randn(50)
z = np.random.randn(50)
color = np.random.randn(50)

fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=8,
        color=color,
        colorscale='Viridis',
        opacity=0.8
    )
)])


fig.update_layout(
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    ),
    title='3D Scatter Plot Example'
)

fig.show()
```
